```python
from skimage import io
from pathlib import Path, PurePath
import numpy as np
```

**reading in and adding noise**

```python
in_dir = Path('data/train_Pt_16/30um/image')
out_dir = Path('data/train_Pt_16_noisy/30um/image')
stdev = 11.88

for path in in_dir.glob('*'):
    print(path)
    img = io.imread(path)
    noise = np.random.normal(0, stdev, (256, 256))
    noisy_img =img+noise
    noisy_img=np.clip(noisy_img, 0, 255)
    io.imsave(out_dir / PurePath(path).name, noisy_img.astype('uint8'))
```
