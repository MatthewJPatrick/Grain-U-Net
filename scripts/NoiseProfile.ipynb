```python
from skimage import io
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = '/content/drive/MyDrive/{}/data/noise_profile/noise.PNG'.format(FOLDERNAME)

img = io.imread(path, as_gray=True)
io.imshow(img)

```

```python
histogram, bin_edges = np.histogram(img, bins=256, range=(0, 1))

plt.figure()
plt.xlim([0.0, 1])
plt.plot(bin_edges[0:-1], histogram)
plt.show()
#plt.savefig('noise_profile.png')
```

```python
std = np.std(img)
print(std)
```
